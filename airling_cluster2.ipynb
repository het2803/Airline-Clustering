# Step 1: Import libraries and dataset #

```python
# Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import os
import openpyxl
import seaborn as sns
```

# Step 2: Load Dataset File #

```python
# use glob to get all the csv files
# in the folder
path = os.getcwd()

# Import Dataset
aline=pd.read_excel(path + "\Data\EastWestAirlines.xlsx", sheet_name='data')
aline
```

# Step 3: Clean Dataset #

```python
aline2=aline.drop(['ID#'],axis=1)
aline2
```

```python
aline.info()
```

```python
# Dataset column renaming

aline.columns = ['ID','Balance', 'Qual_miles', 'cc1_miles', 'cc2_miles', 'cc3_miles','Bonus_miles', 'Bonus_trans', 
                 'Flight_miles_12mo', 'Flight_trans_12','Days_since_enroll', 'Award'] 

# use aline.keys() or aline.columns to get cloumn name list
aline.info()
```

# Step 4: Data Processing & Reporting #

```python
# Normalize heterogenous numerical data by using Standard Scaler
aline2_norm=StandardScaler().fit_transform(aline2)
```

```python
# Use Elbow Graph to find optimum number of  clusters (K value) from K values range
# The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion WCSS 
# random state can be anything from 0 to 42, but the same number to be used everytime,so that the results don't change.
```

```python
# Previously miles award status

aline['Award'].value_counts().plot(kind='pie', autopct='%2.0f%%', fontsize='18', 
                                        colors = ['#F11A05','#43E206'], shadow =True)
plt.show()
```

```python
# Checking trend between balance miles & no. of days since enroll
import seaborn as sns
fig, ax =plt.subplots(figsize=(40,8))
ax = sns.lineplot(x= 'Days_since_enroll', y='Balance',data = aline)
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(aline2_norm)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Graph')
plt.xlabel('Amount of clusters')
plt.ylabel('Within-Cluster Sum-of-Squares')
plt.show()
```

# Step 5: Build Cluster algorithm using K=4 #

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(aline2_norm)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
air4=aline2.copy()
air4["clusters4id"]=clusters4.labels_
air4
```

```python
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
air4.groupby("clusters4id").agg(["mean"]).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(air4['clusters4id'],air4['Balance'], c=clusters4.labels_) 
```

# Step 6: Build Cluster algorithm using K=5 #

```python
# Cluster algorithm using K=5
clusters5=KMeans(5,random_state=30).fit(aline2_norm)
clusters5
```

```python
clusters5.labels_
```

```python
# Assign clusters to the data set
air5=aline2.copy()
air5['clusters5id']=clusters5.labels_
air5
```

```python
# Compute the centroids for K=5 clusters with 11 variables
clusters5.cluster_centers_
```

```python
# Group data by Clusters (K=5)
air5.groupby('clusters5id').agg(['mean']).reset_index()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(air5['clusters5id'],air5['Balance'], c=clusters5.labels_)
```
